```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

test_data = pd.read_csv('/media/dan/Data/git/spacetimeformer/spacetimeformer/test_data.csv')
test_data1 = pd.read_csv('/media/dan/Data/git/spacetimeformer/spacetimeformer/test_data1.csv')
train_data = pd.read_csv('/media/dan/Data/git/spacetimeformer/spacetimeformer/train_data.csv')
train_data1 = pd.read_csv('/media/dan/Data/git/spacetimeformer/spacetimeformer/train_data1.csv')
val_data = pd.read_csv('/media/dan/Data/git/spacetimeformer/spacetimeformer/val_data.csv')
val_data1 = pd.read_csv('/media/dan/Data/git/spacetimeformer/spacetimeformer/val_data1.csv')
```

```python
def plot_vs(a,b, context=None):

    if context is None:
        context = len(a)
    else:
        #make sure + context doesn't go out of bounds

        random_start = np.random.randint(0, len(a)-context)
        if random_start + context > len(a):
            random_end = len(a)
        else:
            random_end = random_start + context
        a = a[random_start:random_end]
        b = b[random_start:random_end]
    
    flat_a = a['flat']
    flat_b = b['flat']
    linear_a = a['linear']
    linear_b = b['linear']
    exp_a = a['exp-decrease']
    exp_b = b['exp-decrease']
    #plot a on one y axis and b on another. make 3 plots in one figure
    fig, ax = plt.subplots(1,3,figsize=(20,5))
    ax[0].plot(flat_a, label='flat', marker='.', markersize=1)
    ax2 = ax[0].twinx()
    ax2.plot(flat_b, label='flat', color='orange', marker='.', linestyle='--', markersize=1)
    ax[0].set_title('Flat')
    ax[0].grid()

    ax[1].plot(linear_a, label='linear', marker='.', markersize=1)
    ax2 = ax[1].twinx()
    ax2.plot(linear_b, label='linear', color='orange', marker='.', linestyle='--', markersize=1)
    ax[1].set_title('Linear')
    ax[1].grid()

    ax2 = ax[2].twinx()
    ax[2].plot(exp_a, label='exp-decrease', marker='.', markersize=1)
    ax2.plot(exp_b, label='exp-decrease', color='orange', marker='.', linestyle='--', markersize=1)
    ax[2].set_title('Exponential')
    ax[2].grid()
    plt.show()
```

```python
plot_vs(test_data, test_data1,10)
```

```python
plot_vs(test_data, train_data,10)
```
